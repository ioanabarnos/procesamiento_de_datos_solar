```python
# Y un nuevo directorio dentro de este para guardar los datos dentro
if not os.path.isdir('data/'):  #si no existe el directorio 'data/' dentro del directorio 'solar_dataset'
    os.mkdir('data/')   #créalo
```

```python
#  Definimos la variable URL con el enlace que utilizaremos para descargar los datos1
URL_data = 'https://archive.ics.uci.edu/ml/machine-learning-databases/solar-flare/flare.data1' 

# Le indicamos que descargue el contenido de la URL como 'response' para más eficiencia
with requests.get(URL_data) as response:
    open(os.path.join(os.getcwd(),'data\\solar.data1'), "wb").write(response.content)
```

```python
#  Definimos la variable URL con el enlace que utilizaremos para descargar los datos2
URL_data = 'https://archive.ics.uci.edu/ml/machine-learning-databases/solar-flare/flare.data2' 

# Le indicamos que descargue el contenido de la URL como 'response' para más eficiencia
with requests.get(URL_data) as response:
    open(os.path.join(os.getcwd(),'data\\solar.data2'), "wb").write(response.content)
```

```python
#  Definimos la variable URL con el enlace que utilizaremos para descargar la metadata 
URL_names = 'https://archive.ics.uci.edu/ml/machine-learning-databases/solar-flare/flare.names' 

# Le indicamos que descargue el contenido de la URL como 'metadata' para más eficiencia
with requests.get(URL_names) as metadata:
    open(os.path.join(os.getcwd(), 'data\\solar.names'), "wb").write(metadata.content)
```

```python
# Leemos datos
## Leemos primer fichero de datos
with open(os.path.join(os.getcwd(),'data\\solar.data1'),'r') as f:
    data1 = f.read().splitlines() # Dividimos el texto por saltos de línea
    data1 = [elem.split(' ') for elem in data1 if elem!=''] # Dividimos cada línea por las comas y removemos líneas vacías
    data1 = data1[1:] # Quitamos la línea de metadata temporal
## Leemos segundo fichero de datos
with open(os.path.join(os.getcwd(),'data\\solar.data2'),'r') as f:
    data2 = f.read().splitlines() # Dividimos el texto por saltos de línea
    data2 = [elem.split(' ') for elem in data2 if elem!=''] # Dividimos cada línea por las comas y removemos líneas vacías
    data2 = data2[1:] # Quitamos la línea de metadata temporal
## Combinamos ambas listas
data = data1+data2
```

```python
# Leemos metadata
with open(os.path.join(os.getcwd(),'data\\solar.names'),'r') as f:
    metadata = f.read().splitlines()
```

```python
metadata
```

```python
## Regex
regex_fn = lambda text: re.findall('^\s+[0-9]+\.{1}\s{1}[a-zA-Z- ]+', text)
reg_text_fn = lambda text : re.findall('[a-zA-Z-]+', text)
metadata_list = [regex_fn(elem)[0].strip() for elem in metadata if regex_fn(elem)]
col_names = [reg_text_fn(elem)[0] for elem in metadata_list if reg_text_fn(elem)]
```

```python
# Construimos el objeto pd.DataFrame
df = pd.DataFrame(data=data, columns=col_names)
```
